# Korelace podílu prefenčních hlasů pro kandidátky ANO a indexů pro obce v jednotlivých krajích

Korelace indexů s podíly prefenčních hlasů v obcích v daných krajích. Vždy pro každou stranu zvlášť.

```python
import pandas as pd

# obce_kraje
obce_kraje = pd.read_csv("obce_kraje.csv")
obce_kraje = obce_kraje.rename(columns={"OBEC_PREZ": "KodObec"})

#výsledky voleb - okrsky pro dopočítání celkového počtu hlasů pro stranu za obec
vysledky = pd.read_csv("2025_ps_vysledky_okrsky.csv")
vysledky = vysledky.rename(columns={"OBEC": "KodObec"})
hlasy_strany = vysledky.groupby(["KodObec", "KSTRANA"])["POC_HLASU"].sum().reset_index()
hlasy_strany = hlasy_strany.rename(columns={"POC_HLASU": "hlasy_strany_celkem"})
#hlasy_strany.to_csv("hlasy_strany_obce.csv")
```

```python
# ANO - kandidátky s aspoň jednou ženou

strany_s_kandidatkami = pd.read_csv("strany_s_kandidatkami_obec.csv")
ANO = strany_s_kandidatkami[strany_s_kandidatkami["KSTRANA"] == 22]

# merge s celkovým počtem hlasů pro stranu + výpočet podílu preferenčních hlasů
ANO = ANO.merge(hlasy_strany, on=["KodObec", "KSTRANA"], how="left")
ANO["podil_preferencnich_z_hlasu"] = ANO["Preferencni_hlasy_celkem_obce"] / ANO["hlasy_strany_celkem"]
```

```python
# 2 - Středočeský kraj
obce_kraj2 = obce_kraje[obce_kraje["VOLKRAJ"] == 2]["KodObec"].tolist()
ANO_kraj2 = ANO[ANO["KodObec"].isin(obce_kraj2)]

ANO_kraj2[['Podil_preferencni_hlasy_zeny_obce', 'Preferencni_hlasy_celkem_obce', 'podil_preferencnich_z_hlasu', 'index_vybavenosti', 'index_socioekonomicky', 'index_stari', 'index_zavisloti', 'index_zavisloti_2', 'index_ekonom_zatizeni', 'index_vzdelani', 'kvartal']].corr()
```
