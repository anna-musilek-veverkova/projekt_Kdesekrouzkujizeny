# Korelace podílu prefenčních hlasů pro kandidátky SPOLU a indexů pro obce v jednotlivých krajích

Korelace indexů s podíly prefenčních hlasů v obcích v daných krajích. Vždy pro každou stranu zvlášť.

```python
import pandas as pd

# obce_kraje
obce_kraje = pd.read_csv("obce_kraje.csv")
obce_kraje = obce_kraje.rename(columns={"OBEC_PREZ": "KodObec"})

# hlasy stran za obec
hlasy_strany = pd.read_csv("hlasy_strany_obce.csv")
```

```python
# Spolu - kandidátky s aspoň jednou ženou

strany_s_kandidatkami = pd.read_csv("strany_s_kandidatkami_obec.csv")
Spolu = strany_s_kandidatkami[strany_s_kandidatkami["KSTRANA"] == 11]

# merge s celkovým počtem hlasů pro stranu + výpočet podílu preferenčních hlasů
Spolu = Spolu.merge(hlasy_strany, on=["KodObec", "KSTRANA"], how="left")
Spolu["podil_preferencnich_z_hlasu"] = Spolu["Preferencni_hlasy_celkem_obce"] / Spolu["hlasy_strany_celkem"]
```

```python
# 2 - Středočeský kraj
obce_kraj2 = obce_kraje[obce_kraje["VOLKRAJ"] == 2]["KodObec"].tolist()
Spolu_kraj2 = Spolu[Spolu["KodObec"].isin(obce_kraj2)]

Spolu_kraj2[['Podil_preferencni_hlasy_zeny_obce', 'Preferencni_hlasy_celkem_obce', 'podil_preferencnich_z_hlasu', 'index_vybavenosti', 'index_socioekonomicky', 'index_stari', 'index_zavisloti', 'index_zavisloti_2', 'index_ekonom_zatizeni', 'index_vzdelani', 'kvartal']].corr()
```
