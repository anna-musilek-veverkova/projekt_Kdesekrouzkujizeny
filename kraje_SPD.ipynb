# Korelace podílu prefenčních hlasů pro kandidátky SPD a indexů pro obce v jednotlivých krajích

Korelace indexů s podíly prefenčních hlasů v obcích v daných krajích. Vždy pro každou stranu zvlášť.

```python
import pandas as pd

# obce_kraje
obce_kraje = pd.read_csv("obce_kraje.csv")
obce_kraje = obce_kraje.rename(columns={"OBEC_PREZ": "KodObec"})

# hlasy stran za obec
hlasy_strany = pd.read_csv("hlasy_strany_obce.csv")
```

```python
# SPD - kandidátky s aspoň jednou ženou

strany_s_kandidatkami = pd.read_csv("strany_s_kandidatkami_obec.csv")
SPD = strany_s_kandidatkami[strany_s_kandidatkami["KSTRANA"] == 6]

# merge s celkovým počtem hlasů pro stranu + výpočet podílu preferenčních hlasů
SPD = SPD.merge(hlasy_strany, on=["KodObec", "KSTRANA"], how="left")
SPD["podil_preferencnich_z_hlasu"] = SPD["Preferencni_hlasy_celkem_obce"] / SPD["hlasy_strany_celkem"]
```

```python
# 2 - Středočeský kraj
obce_kraj2 = obce_kraje[obce_kraje["VOLKRAJ"] == 2]["KodObec"].tolist()
SPD_kraj2 = SPD[SPD["KodObec"].isin(obce_kraj2)]

SPD_kraj2[['Podil_preferencni_hlasy_zeny_obce', 'Preferencni_hlasy_celkem_obce', 'podil_preferencnich_z_hlasu', 'index_vybavenosti', 'index_socioekonomicky', 'index_stari', 'index_zavisloti', 'index_zavisloti_2', 'index_ekonom_zatizeni', 'index_vzdelani', 'kvartal']].corr()
```
