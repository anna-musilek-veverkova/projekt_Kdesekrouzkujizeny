## Pearsonův korelační koeficient
preferenční hlasy pro ženy + charakteristiky obcí

knihovna - seaborn

metoda .corr()

```python
import pandas as pd
import seaborn as sns
data = pd.read_csv("Porovnavaci_tabulka.csv")
korelace = data.corr()
korelace.to_csv("korelacnimatice.csv")
```

## Pearsonův korelační koeficient
preferenční hlasy pro ženy + charakteristiky obcí - pouze 4 kvartil (nejvyšší podíl preferenčních hlasů)

```python
import pandas as pd
import seaborn as sns
data = pd.read_csv("Porovnavaci_tabulka.csv")
data = data[data["kvartil"]==4]
korelace = data.corr()
korelace.to_csv("korelacnimatice4kvartal.csv")
```

# Metoda lineární regrese
knihovna - statsmodels.formula.api

metoda - smf.ols

```python
import pandas as pd
df = pd.read_csv('Preferencni_hlasy_strany_typ.csv')
import statsmodels.formula.api as smf
model = smf.ols('Podil_preferencni_hlasy_zeny_strany ~ C(Strany_typ)', data=df).fit()
print(model.summary())
```

# Vizualizace regrese

knihovna - Matplotlib




```python
import seaborn as sns
import matplotlib.pyplot as plt
df_bloky = pd.read_csv('Preferencni_hlasy_strany_typ.csv')
plt.figure(figsize=(8, 6))

ax = sns.barplot(
    data=df_bloky, 
    x='Strany_typ', 
    y='Podil_preferencni_hlasy_zeny_strany',
    hue='Strany_typ',
    palette= ['#0C293B', '#6C0E0E'],
    legend=False,
    capsize=.1,
    errorbar='sd',
    alpha=0.8
)

sns.swarmplot(
    data=df_bloky, 
    x='Strany_typ', 
    y='Podil_preferencni_hlasy_zeny_strany',
    color="black", 
    alpha=0.5
)

sns.despine()
plt.title("Rozdíl v podílu preferenčních hlasů pro ženy\n(Liberální vs. Konzervativní blok)", fontsize=14, fontweight='bold')
plt.ylabel("Podíl preferenčních hlasů (%)")
plt.xlabel('')
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.show()
```
