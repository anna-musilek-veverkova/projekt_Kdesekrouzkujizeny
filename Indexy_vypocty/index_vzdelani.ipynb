# Index vzdělání 

Je konstruován jako vážený průměr úrovně vzdělání obyvatel obce. Každé kategorii vzdělání je přiřazena váha (základní = 1, střední bez maturity = 2, střední s maturitou/VOŠ = 3, vysokoškolské = 4) a výsledná hodnota odpovídá váženému průměru podílů jednotlivých kategorií. Škála indexu se pohybuje od 1 do 4.

Vytvoření indexu:
- data Sčítání lidu, bytů a domů 2021

Váhové ohodnocení kategorií:
1 bod: Základní vzdělání (včetně bez vzdělání a nedokončeného ZŠ)
2 body: Střední odborné bez maturity (vyučen)
3 body: Úplné střední s maturitou (gymnázia, střední odborné školy) a vyšší odborné vzdělání (VOŠ)
4 body: Vysokoškolské vzdělání (bakalářské, magisterské, doktorské)
Výpočet podílů: Pro každou kategorii vypočítejte její procentuální podíl na celkovém počtu obyvatel. Výsledek vyjádřete jako desetinné číslo (např. 30 % = 0,30).

I = (P_ZS * 1) + (P_SSbez * 2) + (P_SSmat * 3) + (P_VS * 4)

```python
# index vzdělání dle dat SLBD 2021
#  Váhové ohodnocení kategorií
# 1 bod: Základní vzdělání (včetně bez vzdělání a nedokončeného ZŠ)
# 2 body: Střední odborné bez maturity (vyučen)
# 3 bod: Úplné střední s maturitou (gymnázia, střední odborné školy) a vyšší odborné vzdělání (VOŠ)
# 4 body: Vysokoškolské vzdělání (bakalářské, magisterské, doktorské)
# Výpočet podílů: Pro každou kategorii vypočítejte její procentuální podíl na celkovém počtu obyvatel. Výsledek vyjádřete jako desetinné číslo (např. 30 % = 0,30).

# I = (P_ZS * 1) + (P_SSbez * 2) + (P_SSmat * 3) + (P_VS * 4)

import pandas as pd

edu = pd.read_csv("2021_SLDB_OBYVATELSTVO.CSV", encoding="cp1250", usecols=[
"uzkod"
,"vse2121"
,"vse2131"
,"vse2141"
,"vse2151"
,"vse2161"
,"vse2171"
,"vse2181"
,"vse4111"])

edu["P_ZS"] = ((edu["vse2121"] + edu["vse2131"]) / edu["vse4111"])*1
edu["P_SSbez"] = (edu["vse2141"] / edu["vse4111"])*2
edu["P_SSmat"] = ((edu["vse2151"] + edu["vse2161"] + edu["vse2171"] ) / edu["vse4111"])*3
edu["P_VS"] = (edu["vse2181"] / edu["vse4111"])*4

edu["index_vzdelani"] = (edu["P_ZS"] + edu["P_SSbez"] + edu["P_SSmat"] + edu["P_VS"])

edu = edu.rename(columns={"uzkod": "KodObec"})

# Hodnota 100 by znamenala populaci bez vzdělání, hodnota 400 populaci se 100% podílem vysokoškoláků.
#edu["index_vzdelani"] = (edu["P_ZS"] + edu["P_SSbez"] + edu["P_SSmat"] + edu["P_VS"])*100

edu = edu.drop(columns=
["vse2121"
,"vse2131"
,"vse2141"
,"vse2151"
,"vse2161"
,"vse2171"
,"vse2181"
,"vse4111"
,'P_ZS', 'P_SSbez', 'P_SSmat', 'P_VS'])

edu.to_csv("index_vzdelani.csv", index=False)
```
