Index věkové struktury

Věková struktura obce je zachycena čtyřmi standardními demografickými ukazateli vypočítanými na základě metodiky demografické analýzy :
index stáří vyjadřuje počet osob ve věku 65 a více let připadající na 100 dětí ve věku 0–14, kdy hodnoty pod 100 signalizují převahu seniorů nad dětmi. Hodnoty indexu pod 50 znamenají velmi mladou nebo produktivní populaci (málo dětí i seniorů), hodnoty v rozmezí 50–70 jsou typické pro průměrnou českou obec a hodnoty nad 100 znamenají, že neproduktivní složka převyšuje tu produktivní (hodně seniorů + dětí).
Index ekonomického zatížení (závislosti) je poměr neproduktivní složky (děti+senioři) k produktivní složce populace (15–64 let) 
Index závislosti I je počet osob ve věku 65 a více let na 100 dětí (0−14 let)
Index závislosti II je počet osob postproduktivního věku (60+ či 65+) na 100 osob v produktivním věku

Více o indexu viz https://geography.upol.cz/soubory/lide/halas/SGV/06_vekova%20struktura,%20prognozy.pdf 

Vytvoření indexu:
- data PAQ Research (2024) - věkové skupiny 0-14let, 15-64let, 65+let, celkový počet obyvatel 



```python
#index věkové struktury

import pandas as pd

paq_vek = pd.read_csv("dataPAQ_upravena/data_paq_joined.csv")

paq_vek = paq_vek.rename(columns={"Kod_obce": "KodObec"})
paq_vek["KodObec"] = paq_vek['KodObec'].astype(str)
sloupce_obj = [c for c in paq_vek.select_dtypes(include="str").columns if c != "KodObec"]
paq_vek[sloupce_obj] = paq_vek[sloupce_obj].replace("data chybí", pd.NA)

paq_vek['VEK_65AVIC'] = paq_vek['VEK_65AVIC'].astype("Int64")
paq_vek['VEK_15_64'] = paq_vek['VEK_15_64'].astype("Int64")
paq_vek['VEK_0_14'] = paq_vek['VEK_0_14'].astype("Int64")
paq_vek['OBYVATELSTVO_POCET_2024'] = paq_vek['OBYVATELSTVO_POCET_2024'].astype("Int64")

# podíly věkových skupin
paq_vek["podil_mladi"] = paq_vek["VEK_0_14"]/paq_vek["OBYVATELSTVO_POCET_2024"] # 0-14 let
paq_vek["podil_produktivni"] =  paq_vek["VEK_15_64"]/paq_vek["OBYVATELSTVO_POCET_2024"] # 15-64 let
paq_vek["podil_65plus"] = paq_vek["VEK_65AVIC"]/paq_vek["OBYVATELSTVO_POCET_2024"] # 65+ let

#Index stáří: Udává počet osob ve věku 65 a více let na 100 dětí (0–14 let). Pokud je hodnota vyšší než 100, znamená to, že v populaci převažují senioři nad dětmi.
paq_vek["index_stari"] = (paq_vek["VEK_0_14"] / paq_vek["VEK_65AVIC"])*100

#Index závislosti I: Počet dětí (0–14 let) na 100 osob v produktivním věku (15–59 let nebo 15–64 let).
paq_vek["index_zavisloti"] = (paq_vek["VEK_0_14"] / paq_vek["VEK_15_64"])*100

#Index závislosti II: Počet osob poproduktivního věku (60+ či 65+) na 100 osob v produktivním věku.
paq_vek["index_zavisloti_2"] = (paq_vek["VEK_65AVIC"] / paq_vek["VEK_15_64"])*100

#Index ekonomického zatížení: Celkový počet dětí a seniorů závislých na 100 osobách v produktivním věku
paq_vek["index_ekonom_zatizeni"] = ((paq_vek["VEK_0_14"] + paq_vek["VEK_65AVIC"]) / paq_vek["VEK_15_64"])*100

paq_vek = paq_vek.drop(columns=['Nazev_obce', 'Kod_okresu', 'Nazev_okresu', 'Kod_kraje',
       'Nazev_kraje', 'Prispevky_bydleni_avgMonth_2024',
       'Doplatky_bydleni_avgMonth_2024', 'Pridavky_deti_avgMonth_2024',
       'Prispevky_pece_avgMonth_2024', 'Prispevky_zivobyti_avgMonth_2024',
       'NEZAMESTNANOST', 'Dlouhodoba_nezamestnanost_2024', 'PECOVATELAK',
       'STREDISKO_VOLNY_CAS', 'HUSTOTA_ZALIDNENI','Role_obce_na_predskolni_peci_2021','DETSKE_SKUPINY', 'DETSKE_HRISTE', 'KULTURAK', 'KINO',
       'POCET_NAROZENYCH_DETI', 'MUZI_SAMOZIVITELE', 'ZENY_SAMOZIVITELKY',
       'DETI_3_4_BEZ_DOCHAZKY_DO_MS', 'LIDE_EXEKUCE',
       'KOSTEL', 'OBECNI_POLICIE',"VEK_65AVIC",'VEK_15_64','VEK_0_14','OBYVATELSTVO_POCET_2024', 'podil_mladi', 'podil_produktivni','podil_65plus'])

paq_vek.to_csv("index_vekove_struktury.csv", index=False)
```
